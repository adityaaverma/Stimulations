```python
def gram_schmidt(a,b,c):
  p=norm(a)
  q1=[x/p for x in a]

  # Calculate q2_raw = b - proj(b, q1)
  proj_b_q1 = projection([q1], b)
  q2=[b_val - p_val for b_val, p_val in zip(b, proj_b_q1)]
  p=norm(q2)
  q2=[x/p for x in q2]

  # Calculate q3_raw = c - proj(c, q1) - proj(c, q2)
  proj_c_q1 = projection([q1], c)
  proj_c_q2 = projection([q2], c)
  q3=[c_val - p1_val - p2_val for c_val, p1_val, p2_val in zip(c, proj_c_q1, proj_c_q2)]
  p=norm(q3)
  q3=[x/p for x in q3]
  return q1,q2,q3

a=[1, 1, 2]
b=[1, -1, 0]
c=[1, 0, 4]
x=[gram_schmidt(a,b,c)]
print(x)
```
