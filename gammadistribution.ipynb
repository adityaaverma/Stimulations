# Monte Carlo Estimation of n-Dimensional Unit Ball Volume

This notebook demonstrates how to use Monte Carlo simulation to estimate the volume of the unit ball in n dimensions. The volume of an n-dimensional unit ball is given by:

$$V_n = \frac{\pi^{n/2}}{\Gamma(n/2 + 1)}$$

We'll use random sampling to approximate this volume and visualize the results.

```python
def stimulate(n,trials=100000):
    points=np.random.uniform(-1,1,(trials,n))
    inside=np.sum(np.sum(points**2,axis=1)<=1)
    
    return inside/trials

```

## Step 1: Define the Monte Carlo Simulation Function

The `stimulate` function performs Monte Carlo integration to estimate the volume of the n-dimensional unit ball. It generates random points uniformly distributed in the [-1,1]^n hypercube and counts how many fall inside the unit ball (where the Euclidean norm ≤ 1). The ratio of points inside to total points approximates the volume ratio.

```python
results={}
for n in range(1,11):
    results[n]=stimulate(n)
```

## Step 2: Run Simulations for Multiple Dimensions

We run the Monte Carlo simulation for dimensions n=1 through n=10, storing the estimated probabilities (which are proportional to the volume ratios) in a dictionary.

```python

def plot_results_interactive(results):
    n_vals = list(results.keys())
    probs = list(results.values())

    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=n_vals,
        y=probs,
        mode='lines+markers',
        name='Probability'
    ))

    fig.update_layout(
        title="Probability that point lies in unit ball vs dimension",
        xaxis_title="n",
        yaxis_title="Probability",
        hovermode="x unified"
    )

    fig.update_layout(
    yaxis_type="log"
    )

    fig.show()
```

## Step 3: Define Interactive Plotting Function

We create an interactive Plotly visualization to show how the estimated probability (volume ratio) changes with dimension. The plot uses a logarithmic y-axis to better show the rapid decrease in volume as dimensions increase.

```python
# plot_results(results)              # static
plot_results_interactive(results) # interactive
```

## Step 4: Display the Interactive Plot

Execute this cell to generate the interactive Plotly chart showing the relationship between dimension and volume estimation.

```python
import matplotlib.pyplot as plt
import math
from ipywidgets import interact, IntSlider


def plot_volume_estimates(n=3, trials=100000, repeats=10):
    estimates = [stimulate(n, trials) for _ in range(repeats)]
    true_volume = np.pi**(n/2) / ((2**n) *math.gamma(n/2 + 1))

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.hist(estimates, bins=12, alpha=0.8, edgecolor='black')
    ax.axvline(true_volume, color='red', linewidth=2, label=f'True volume = {true_volume:.4f}')
    ax.set_title(f'Monte Carlo volume estimates for n={n} with {trials} samples')
    ax.set_xlabel('Estimated volume')
    ax.set_ylabel('Frequency')
    ax.legend()
    plt.show()


@interact(
    n=IntSlider(value=3, min=1, max=10, step=1, description='Dimension'),
    trials=IntSlider(value=10000, min=1000, max=50000, step=1000, description='Trials'),
    repeats=IntSlider(value=10, min=1, max=50, step=1, description='Repeats')
)
def interactive_plot(n, trials, repeats):
    plot_volume_estimates(n, trials, repeats)
```
