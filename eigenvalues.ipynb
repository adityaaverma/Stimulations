```python
import numpy as np
```

```python
A=np.array([[0.8,0.3],
           [0.2,0.7]],dtype=float)

x=np.array([1.0,1.0])
x=x/np.linalg.norm(x)

for k in range(1000):
    x=A @ x
    x=x/np.linalg.norm(x)

    print(f"iter {k+1}: {x}")
```

```python


# Compute eigenvalues and eigenvectors
eigenvals, eigenvecs = np.linalg.eig(A)
print("Eigenvalues:", eigenvals)
print("Eigenvectors:")
print(eigenvecs)

# Dominant eigenvalue and eigenvector
dominant_idx = np.argmax(np.abs(eigenvals))
dominant_eigenval = eigenvals[dominant_idx]
dominant_eigenvec = eigenvecs[:, dominant_idx]
print(f"\nDominant eigenvalue: {dominant_eigenval}")
print(f"Dominant eigenvector: {dominant_eigenvec}")

# Check if your iterated x matches
print(f"\nYour final x: {x}")
print(f"Matches dominant eigenvector: {np.allclose(x, dominant_eigenvec / np.linalg.norm(dominant_eigenvec))}")
```

```python
eig_A=np.linalg.eigvals(A)
eig_A_plus_I=np.linalg.eigvals(A + np.eye(A.shape[0]))
eig_A_sq=np.linalg.eigvals(A @ A)
eig_A_inv=np.linalg.eigvals(np.linalg.inv(A))

print("Eigenvalues of A:      ", eig_A)
print("Eigenvalues of A + I:  ", eig_A_plus_I)
print("Eigenvalues of A^2:    ", eig_A_sq)
print("Eigenvalues of A^-1:   ", eig_A_inv)

```

```python
Ak=np.eye(A.shape[0])
for k in range(1,11):
    Ak=Ak @ A
    print(f"\nA^{k}=\n{Ak}")
```
